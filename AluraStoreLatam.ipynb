### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

#1. Análisis de facturación



```python

#Identificando la ganancia neta de cada tienda

ganancia_tienda1_neto = tienda['Precio'].sum()
ganancia_tienda2_neto = tienda2['Precio'].sum()
ganancia_tienda3_neto = tienda3['Precio'].sum()
ganancia_tienda4_neto = tienda4['Precio'].sum()

#Identificando costos de envío

costo_envio_tienda = tienda['Costo de envío'].sum()
costo_envio_tienda_2 = tienda2['Costo de envío'].sum()
costo_envio_tienda_3 = tienda3['Costo de envío'].sum()
costo_envio_tienda_4 = tienda4['Costo de envío'].sum()

#Ganancia real por Tienda, se hará la diferencia entre la ganancia neta y sus costos de envío

ganancia_real_tienda = ganancia_tienda1_neto - costo_envio_tienda
ganacia_real_tienda_2 = ganancia_tienda2_neto - costo_envio_tienda_2
ganancia_real_tienda_3 = ganancia_tienda3_neto - costo_envio_tienda_3
ganancia_real_tienda_4 = ganancia_tienda4_neto - costo_envio_tienda_4

#Aplicando lo aprendido

lista_tienda = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
lista_ganancia_final = [ganancia_real_tienda, ganacia_real_tienda_2, ganancia_real_tienda_3, ganancia_real_tienda_4]
colors = ['gray', 'brown', 'green', 'blue']
for i in range(len(lista_tienda)):
  print(f'Ganancia de {lista_tienda[i]} es: {lista_ganancia_final[i]}')
```

```python
#Gráfico de la Primera Consulta
plt.title('Ganancia por tienda (G.1)')
plt.xlabel('Ganancia')
plt.ylabel('Tiendas')
plt.barh(y= lista_tienda, width= lista_ganancia_final, color= colors)
plt.xlim(850000000, 1100000000)
plt.show()
```

```python
#Se aprecia en el Gráfico que la Tienda 1 es la que tiene una mayor ganancia real.
#En cambio la Tienda 4 la menor ganancia real.
```

# 4. Productos más y menos vendidos

```python
#lista en string
tiendas_str = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#lista con la data de cada tienda
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

#Identificar tipo de producto de manera unica
nombre_productos_unicos = tienda['Producto'].unique()

#Ordenar la lista
nombre_productos_unicos.sort()

dicc = {}
dicc["Producto"] = nombre_productos_unicos

# crea una lista de las cantidades de productos vendidos por tipo en cada tienda
cantidad_productos_vendidos = []
lista_auxiliar = []

#Con doble método for se crea la manera de almacenar la data
for una_tienda in lista_tiendas:
  for producto_n in nombre_productos_unicos:
    lista_auxiliar.append(list(una_tienda["Producto"]).count(producto_n))
  cantidad_productos_vendidos.append(lista_auxiliar)
  lista_auxiliar = []

#Diccionario con la información de venta de cada tienda
tabla_cantidad_productos = {llaves:valores for (llaves, valores) in zip(tiendas_str, cantidad_productos_vendidos)}

#A lo anterior se le adiciona los nombres de los productos
tabla_cantidad_productos = {**dicc, **tabla_cantidad_productos}


#A la información anterior se le crea su tabla(DataFrame)
tabla_productos = pd.DataFrame(tabla_cantidad_productos)
#Completado
tabla_productos
```

```python
#Crear un espacio de memoria enfocado a la tienda 1
#Obteniendo los 5 Productos más y menos vendidos

tabla_productos_tienda_1 = tabla_productos.sort_values(by = "Tienda 1")
productos_mas_vendidos_t1 = tabla_productos_tienda_1.tail(5)
productos_menos_vendidos_t1 = tabla_productos_tienda_1.head(5)

#Crear un espacio de memoria enfocado a la tienda 2
#Obteniendo los 5 Productos más y menos vendidos

tabla_productos_tienda_2 = tabla_productos.sort_values(by = "Tienda 2")
productos_mas_vendidos_t2 = tabla_productos_tienda_2.tail(5)
productos_menos_vendidos_t2 = tabla_productos_tienda_2.head(5)

#Crear un espacio de memoria enfocado a la tienda 3
#Obteniendo los 5 Productos más y menos vendidos

tabla_productos_tienda_3 = tabla_productos.sort_values(by = "Tienda 3")
productos_mas_vendidos_t3 = tabla_productos_tienda_3.tail(5)
productos_menos_vendidos_t3 = tabla_productos_tienda_3.head(5)

#Crear un espacio de memoria enfocado a la tienda 4
#Obteniendo los 5 Productos más y menos vendidos

tabla_productos_tienda_4 = tabla_productos.sort_values(by = "Tienda 3")
productos_mas_vendidos_t4 = tabla_productos_tienda_4.tail(5)
productos_menos_vendidos_t4 = tabla_productos_tienda_4.head(5)
```

# 6. Información adicional: Ventas distribuídas por ciudad
